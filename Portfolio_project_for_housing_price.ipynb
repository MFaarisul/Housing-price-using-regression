# Handling Missing Values

```python
import seaborn as sns
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cbar=False)
```

```python
#drop the url columns
df.drop(['url', 'region_url', 'image_url'], axis=1, inplace=True)
```

```python
df['laundry_options'].value_counts()
```

```python
#check the most common laundry option for each type of the house
f_laundry = df.groupby(['type', 'laundry_options']).count().sort_values(['type','beds'], ascending=False)
f_laundry
```

```python
house_type = df['type'].value_counts().index
house_type
```

```python
#generate a function that fill the missing value with the most common laundry option for each type of the house
def laundry_fill(spec):
    ht = spec[0]
    laundry = spec[1]
    
    if pd.isnull(laundry):
        for t in house_type:
            if ht == t:
                return f_laundry.loc[ht].index[0]
    else:
        return laundry
```

```python
#perform the function
df['laundry_options'] = df[['type','laundry_options']].apply(laundry_fill, axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cbar=False)
```

```python
#check the most common parking option for each type of the house
f_parking = df.groupby(['type', 'parking_options']).count().sort_values(['type','beds'], ascending=False)
f_parking
```

```python
#generate a function that fill the missing value with the most common parking option for each type of the house
def parking_fill(spec):
    ht = spec[0]
    parking = spec[1]
    
    if pd.isnull(parking):
        for t in house_type:
            if ht == t:
                return f_parking.loc[ht].index[0]
    else:
        return parking
```

```python
#perform the function
df['parking_options'] = df[['type','parking_options']].apply(parking_fill, axis=1)
```

# Feature Engineering

```python
check_outlier = pd.DataFrame({'Q1':df_final.groupby(['type']).quantile(0.25)['price'],
                              'Q2':df_final.groupby(['type']).median()['price'],
                              'Q3':df_final.groupby(['type']).quantile(0.75)['price']})
```

```python
iqr = check_outlier['Q3'] - check_outlier['Q1']
check_outlier['h_outlier'] = check_outlier['Q3'] + (iqr * 1.5)
check_outlier['l_outlier'] = check_outlier['Q1'] - (iqr * 1.5)
```

```python
check_outlier
```

```python
type_house = check_outlier.index
type_house
```

```python
check_outlier.loc['apartment']
```

```python
def drop_outlier(type_house):
    df_outlier = df_final
    new_df = df_final.drop(df_final.index,axis=0)
    
    for t in type_house:
        house = df_outlier[df_outlier['type'] == t]
        house_final = house[(house['price'] < check_outlier.loc[t][3]) & (house['price'] > check_outlier.loc[t][4])]
        
        new_df = pd.concat([new_df, house_final])
    return new_df
```

```python
df_final = drop_outlier(type_house)
```

```python
df_final.shape
```

```python
df_final['price'].hist()
```

```python
sns.distplot(df_final['price'])
```

```python
sns.boxplot(x=df_final['price'], y=df_final['type'])
```

```python
obj_features = df_final.dtypes[df_final.dtypes == 'object'].index

def dummies(obj_features):
    df_concat = df_final
    
    for feat in obj_features:
        df_dummies = pd.get_dummies(df_final[feat], drop_first=True)
        df_concat = pd.concat([df_concat, df_dummies], axis=1)
        
        df_concat.drop(feat, axis=1, inplace=True) 
        
    return df_concat
```

```python
df_final = dummies(obj_features)
```

# Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
```

```python
X = df_final.drop('price', axis=1)
y = df_final['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
def rmse(model):
    return np.sqrt(-cross_val_score(model, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
```

```python
#model 1
lr = LinearRegression()
rmse(lr).mean()
```

```python
#model 2
lasso = LassoCV()
rmse(lasso).mean()
```

```python
#model 3
ridge = RidgeCV()
rmse(ridge).mean()
```

Since model 3 has the lowest rmse, so we use that model to predict a house price

```python
ridge.fit(X_train, y_train)
pred = ridge.predict(X_test)
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

If we take a look at the average of the price in this particular dataset, it is around 8825 and our rmse is 247. that means our model have an error ~3% which is quite good

```python
coeff = pd.DataFrame(ridge.coef_, X.columns)
coeff.columns = ['Coeff']
coeff.sort_values('Coeff', ascending=False)
```

Interpreting the coefficients:

- Holding all other features fixed, a 1 unit increase in **ca** is associated with an increase of **656.88** dollars price
- Holding all other features fixed, a 1 unit increase in **long island** is associated with an increase of **635.75** dollars price
- etc.
